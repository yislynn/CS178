### 3.1

Let X be the first two columns of iris. Plot the classification boundary for varying values of K = [1, 5, 10, 50] with plotClassify2D. Summarize the pros and cons of increasing K.

```python
iris = np.genfromtxt("data/iris.txt",delimiter=None) # load the data
Y = iris[:,-1]
X = iris[:,0:2]

import mltools as ml

X,Y = ml.shuffleData(X,Y);

Xtr,Xva,Ytr,Yva = ml.splitData(X,Y, 0.8); # split data into 80/20 train/validation

for K in [1,5,10,50]:
    print("classification Boundary for K =", K)
    knn = ml.knn.knnClassify() # create the object and train it
    knn.train(Xtr, Ytr, K) # where K is an integer, e.g. 1 for nearest neighbor prediction
    YvaHat = knn.predict(Xva) # get estimates of y for each data point in Xva

    ml.plotClassify2D( knn, Xtr, Ytr ); # make 2D classification plot with data (Xtr,Ytr)
```

When mapping with K = 1, each data point is its own earest neighbor. Because of this, the boundaries are very specific and fine-grained. Each data point pretty much gets its own patch. While this fits incredibly well for the training data, it is definitely a problem for other data sets that differ. The data that fits in the overlapping area will be miscategorised the most, since it could easily be misclassified due to the presence of a point of a different classification.

With K = 5, the boundaries have shifted qute a bit. Though there are still islands of colors within each other, the data points that are more "irregular" such as the yellow points at about (2.5, 4.9) or (2.8, 5.6)  no longer have their own patches amidst the turquoise. This is better when it comes to classification of datapoints belonging to the turquoise group, since most of those data points are concentrated in that area.

At K = 10, the boundaries have shifted again. This time, the purple has shifted downwards, and interestingly enough, there are more turquoise patches in the yellow area. Since more neighbors are being taken into consideration when calculating the classification boundary, the overlapping turquoise and yellow data points changes the boundaries again. There is enough turquoise that overlaps with the yellow for more turquoise patches to form in the yellow classification area. I think having some turquoise patches will be good because the data clearly has a group of overlapping data. With this degree of overlap, being too strict with the boundary will cause creater misclassification than if the boundries were drawn more clear cut, with with K = 50.

Lastly, with K = 50, the boundaries have been drawn quite clearly. The yellow and turquoise boundaries have been split, and there are no longer any islands in the yellow area. The purple area has also dipped quite a bit on the left side. Though the areas are pretty split, they are a little off; the purple data points don't look like they will dip that low based on the shape of the data points, and turqoise data points may be miscategorized as a purple data point. Additionally, some turquoise data points will be mistcategorised as yellow data points, especially those in the range of 6.5 - 7 and 2.75 - 3.3. Quote a few yellow data points will also be miscategorized as turquoise data points as well. Because of this, it is rather obvious that an issue with having a larger K value is that boundary classifications are too simple. The simplicity is especially harmful for datasets like this one, where the classifications will overlap and categorization is more complex.

In summary, K values that are too small are too complex and fit the training data too well. These boundaries will do poorly in classifying other datasets because they mostly conform to one, and other datas sets will have different variability. Meanwhile, if K values are too big, they may become too simple and the classification will not necessarily fit the data too well. It may also cause issues when it comes to more complex datasets, like those that overlap with each other. A moderate K value is ideal because it should fit the data well, but still be generalizable to other datasets.

### 3.2

Let X be the first two features. Compute the error rate on training and validation data as a function of K = [1, 2, 5, 10, 50, 100, 200].

Plot the results using a semi-log plot (semilogx), with training error in red and validation error in green. Based on these plots, what value of K would you recommend? 

```python
K=[1,2,5,10,50,100,200]
Y = iris[:,-1]
X = iris[:,0:2]

errTrain = [0,0,0,0,0,0,0]
errVal = [0,0,0,0,0,0,0]
X,Y = ml.shuffleData(X,Y);

Xtr,Xva,Ytr,Yva = ml.splitData(X,Y, 0.8); # split data into 80/20 train/validation
trsize = Ytr.shape
vasize = Yva.shape

for i,k in enumerate(K):
    knn = ml.knn.knnClassify()
    knn.train(Xtr, Ytr, k)
    YtrHat = knn.predict(Xtr)
    diff = YtrHat - Ytr
    errTrain[i] = np.sum(np.count_nonzero(diff)) / trsize
    #TODO: repeat prediction / error evaluation for validation data
    YvaHat = knn.predict(Xva)
    diff = YvaHat - Yva
    errVal[i] = np.sum(np.count_nonzero(diff)) / vasize

                       
plt.semilogx(K, errTrain, color="red")
plt.semilogx(K, errVal, color="green")
plt.show()
```

Since the validation error seems to be smallest at around K = 10 I think I would suggest using that as the K value. The results of the test data indicate that the trained model performed rather well on a different dataset.
